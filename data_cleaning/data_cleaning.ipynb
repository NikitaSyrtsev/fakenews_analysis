```python
import pandas as pd
```

```python
news = pd.read_csv('ru_test.csv')
news
```

```python
# Посчитаем количество строк с метками 0 и 1
count_0 = (news['is_fake'] == 0).sum()
count_1 = (news['is_fake'] == 1).sum()

print(f"Количество строк с меткой 0: {count_0}")
print(f"Количество строк с меткой 1: {count_1}")
```

```python
news2 = pd.read_csv('ru_train.csv')
news2
```

```python
# Посчитаем количество строк с метками 0 и 1
count_0 = (news2['is_fake'] == 0).sum()
count_1 = (news2['is_fake'] == 1).sum()

print(f"Количество строк с меткой 0: {count_0}")
print(f"Количество строк с меткой 1: {count_1}")
```

```python
# Дублируем столбец title в новый столбец text
news2['text'] = news2['title']

# Меняем порядок столбцов
news2 = news2[['title', 'text', 'is_fake']]

# Сохраняем изменённый датафрейм в новый файл
news2.to_csv('ru_train_with_text.csv', index=False)

# Фильтруем строки, где is_fake == 1
fake_df = news2[news2['is_fake'] == 1]

# Сохраняем отфильтрованные данные в отдельный файл
fake_df.to_csv('final_dataset/final_ru_f2879.csv', index=False)
```

```python
news3 = pd.read_csv('en_t3000_f3000.csv')
news3
```

```python
# Оставляем только нужные столбцы
news3 = news3[['title', 'text', 'label']]

# Переименовываем столбец label в is_fake
news3 = news3.rename(columns={'label': 'is_fake'})

# Заменяем значения FAKE на 1, REAL на 0
news3['is_fake'] = news3['is_fake'].replace({'FAKE': 1, 'REAL': 0})

# Сохраняем результат в новый файл
news3.to_csv('final_dataset/final_en_t3000_f3000.csv', index=False)
```
